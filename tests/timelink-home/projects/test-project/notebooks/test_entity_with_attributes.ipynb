# Usages of entities_with_attribute

## Every person with a given attribute type

For every person with attribute jesuita-entrada show birth place and nationality

```python
from timelink.pandas.group_attributes import group_attributes
from timelink.pandas import entities_with_attribute

# Get list of people with with a certain value in a specific attribute
df = entities_with_attribute(
    entity_type='person',
    show_elements=['name','groupname'],
    the_type='jesuita-entrada',
    more_attributes=['nacionalidade','nascimento'],
    db=tlnb.db,
)
df.info()
ids = list(df.index.unique())
print(len(ids),ids)
# choose the cols to show
cols = ['name','groupname','jesuita-entrada','jesuita-entrada.date','nacionalidade','nascimento', 'nascimento.date']
df[cols].sort_values(['jesuita-entrada','jesuita-entrada.date']).head(50)
```

## entities with a series of attributes



```python
from timelink.pandas.group_attributes import group_attributes
from timelink.pandas import entities_with_attribute

# Get list of people with with a certain value in a specific attribute
df = entities_with_attribute(
    entity_type='person',
    show_elements=['name','groupname'],
    the_type=['nascimento','jesuita-entrada','partida','chegada','estadia','jesuita-votos-local','morte'],
    column_name='lugares',
    more_attributes=[],
    db=tlnb.db,
)
ids = list(df.index.unique())
print(len(ids),ids)
df.reset_index(inplace=True)
df.info()
# choose the cols to show
cols = ['id','name','groupname','lugares.type','lugares','lugares.line','lugares.date']
df[cols].sort_values(['name','id','lugares.line']).head(50)
```
