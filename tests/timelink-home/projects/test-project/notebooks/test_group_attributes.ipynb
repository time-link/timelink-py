```python
from timelink.notebooks import TimelinkNotebook

tlnb = TimelinkNotebook(db_name="group_attribute")
```

```python
# NBVAL_IGNORE_OUTPUT
tlnb.print_info()
```

```python
### Import file
from timelink.api.database import import_from_xml

db = tlnb.db
with db.session() as session:
    import_from_xml( "../../../../xml_data/dehergne-a.xml",session)
```

```python
from timelink.pandas.group_attributes import group_attributes
from timelink.pandas import entities_with_attribute

# Get list of people with with a certain value in a specific attribute
df = entities_with_attribute(
                    entity_type='person',
                    the_type='estadia',
                    the_value='Macau',
                    dates_in=(1580, 1602),
                    db=tlnb.db,
                    sql_echo=False)
ids = list(df.index.unique())
print(len(ids),ids)
group = group_attributes(group=ids,
                         entity_type='person',
                         show_elements=['groupname','name'],
                         db=tlnb.db)
group.info()
```

```python
# NBVAL_IGNORE_OUTPUT

from timelink.pandas.group_attributes import display_group_attributes

display_group_attributes(ids,
                         entity_type='person',
                         # show name,sex,obs
                         header_elements=['name'],
                         # define which attributes to display in the header
                         header_attributes=['nacionalidade','jesuita-entrada'],
                         include_attributes=['estadia','partida'],
                         sort_attributes=['the_date'],
                         db=tlnb.db)
```
