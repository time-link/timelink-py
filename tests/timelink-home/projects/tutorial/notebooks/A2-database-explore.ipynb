# Database explore

## Setup

```python
from timelink.notebooks import TimelinkNotebook

tlnb = TimelinkNotebook()
tlnb.print_info()
```

## Describing the database

### Get list of existing Entity models

```python
models = tlnb.db.get_models_ids()
models
```

### Describe each model

```python
for model in models:
    print()
    tlnb.db.describe(model, show=True)
    print()
```

### Describe tables associated with models

```python
tables = tlnb.db.orm_table_names()
print()
for table in tables:
    print()
    tlnb.db.describe(table, show=True)
```
