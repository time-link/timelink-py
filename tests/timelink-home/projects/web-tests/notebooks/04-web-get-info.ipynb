# Database: get entities

```python
from timelink.api.database import TimelinkDatabase

db_dir = '../database/sqlite/'
db = TimelinkDatabase(db_type='sqlite',
                      db_path=db_dir,
                      db_name='timelink-web')
```

## Get a single entity

```python
from timelink.api.schemas import EntityAttrRelSchema
with db.session() as session:
    mr = db.get_entity('b1685.12-per1', session=session)
    mr_schema = EntityAttrRelSchema.model_validate(mr)

mr_schema.model_dump(exclude=['rels_in'])
```

### Real Entities

```python
from timelink.api.schemas import RealEntitySchema

with db.session() as session:
    re = db.get_entity('rp336670', session=session)
    # can also use RealEntityAttrRelSchema
    re_schema = RealEntitySchema.model_validate(re)
    re_kleio = re.to_kleio()

re_dict = re_schema.model_dump(exclude=['contains'])
```
