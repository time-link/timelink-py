# Import data

Translate new and updated files

```python
for kfile in kserver.translation_status(path='',recurse='yes',status='T'):
    print(kfile.status.value, kfile.path)
    kserver.translate(kfile.path, recurse='no', spawn='no')
```

Import from Kleio Server

```python
import pprint
from timelink.kleio.schemas import KleioFile
from timelink.kleio.importer import import_from_xml

to_import = kserver.translation_status(path='',recurse='yes',status='V')
kfile: KleioFile

with db.session() as session:
    for kfile in to_import:
        print(kfile.status.value, kfile.path)
        try:
            stats = import_from_xml(kfile.xml_url, session=session,
                                    options={"return_stats": True,
                                             "kleio_token": token,
                                             "kleio_url": url,
                                             "mode":"TL"})
            pprint.pprint(stats)
        except Exception as e:
            print(f"Error: {e}")
            continue
```

```python
from timelinknb.pandas import attribute_to_df


# Get list of people with with a certain value in a specific attribute
df = attribute_to_df(
                    the_type='naturalidade',
                    the_value='Coimbra',
                    person_info=True,
                    more_cols=['faculdade','uc-entrada','uc-saida','grau'],
                    sql_echo=False)
df.info()
```

```python
df.head(5)
```



###  Remover colunas sem valores

###  Remove empty columns

```python
df.dropna(how='all', axis=1, inplace=True)
df.info()
```


###  Remover as remissivas

###  Remove cross-references

```python
no_crossref = df['uc-entrada'] != '0000-00-00'
df = df[no_crossref]
```

```python
view_cols = ['name','naturalidade','faculdade','grau','grau.date','uc-entrada','uc-saida']
df[view_cols].head(10)
```

## Counting



###  Contagem de atributos a partir de uma tabela em memória
 
---

###  Count attributes from an existing dataframe



```python
# create a column with the index values which are the id numbers
df['id'] = df.index.values

col = 'faculdade' # subotal by this column

# Use pandas groupby and specify unique value count for id
df_totals = df.groupby(col).agg({'id':'nunique',
                                                  'uc-entrada':'min',
                                                  'uc-saida':'max'})

df_totals.sort_values('id',ascending= False).head(30)
```
